### Reading the NRC VAD Lexicon by Dr. Saif Mohammed

```python
with open('NRC-VAD-Lexicon.txt') as f:
    lines = [line.rstrip().split() for line in f]
  
final_vad = []
for i in range(len(lines)):
    if len(lines[i])>4:
        str1 = " "
        stsr = str1.join(lines[i][0:-3])
        new_l = [stsr] + lines[i][-3:]
        final_vad.append(new_l)
    else:
        final_vad.append(lines[i])
```

```python
dnrc = pd.DataFrame(final_vad)
dnrc.columns = ['Word', 'V','A','D']
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer() #creating an instance of the class
# creating a list of some words to be stemmed
for i in range(len(dsnrc)):
    word = dsnrc.iat[i,0]
    st = ps.stem(word)
    dsnrc.iloc[[i],[0]] = st
```

### Emotion score analysis - NRCLex

```python
df_emotions = rdf_all.copy()
```

```python
df_emotions['s']=0
```

```python
def e_sum(l):
    s=0
    for i in l:
        s+=i[1]
    return s
```

```python
from nrclex import NRCLex
for j in range(len(df_emotions)):
    word = df_emotions.iat[j,0]
    emotion = NRCLex(word)
    df_emotions.iloc[[j],[0]] = word + ' : ' + str(emotion.top_emotions)
    w = e_sum(emotion.top_emotions)
    df_emotions.iloc[[j],[3]] = w
```

```python
df_emotions.to_csv('emotion_raw_nrclex.csv')
```

### Emotion score analysis - SentiWordNet

```python
df_emotions = rdf_all.copy()
```

```python
df_emotions
```

```python
import nltk
from nltk.corpus import wordnet
nltk.download('sentiwordnet')
from nltk.corpus import sentiwordnet as swn

# Function to classify a word as emotion or non-emotion
def classify_word(word):
    synsets = swn.senti_synsets(word)
    pos_score, neg_score = 0, 0
    for synset in synsets:
        pos_score += synset.pos_score()
        neg_score += synset.neg_score()
    if pos_score > 0.5:
        return ["emotion",pos_score,neg_score]
    elif neg_score > 0.5:
        return ["emotion",pos_score,neg_score]
    else:
        return ["neutral",pos_score,neg_score]

# Example usage
words_list2 = list(df_emotions['word'])
df_emotions['s2']=0
df_emotions['pos_score']=0
df_emotions['neg_score']=0
x=0
for j in range(len(df_emotions)):
    word = df_emotions.iat[j,0]
    st = classify_word(word)
    df_emotions.iloc[[j],[4]] = st[1]
    df_emotions.iloc[[j],[5]] = st[2]
    if st[0] =='emotion':
        df_emotions.iloc[[j],[3]] = 1
        
        x+=1
        

    
```
